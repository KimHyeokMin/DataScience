#### 일변량 범주형 자료에 대해 도수분포표(frequency distribution table), 상대도수분포표(relative frequency distribution table)를 사용하여 원자료(raw data)의 정리와 요약

### 1. 도수분포표, 상대도수분포표로 일변량(one variable) 범주형 자료(categorical variable)에 대한 정리와 요약

#### (1) 일변량 도수분포표(frequency distribution table)
> **변량** : 데이터

> **계급** : 변량을 일정한 구간으로 나눈 구간

> **도수** : 각 계급에 속하는 변량의 개수

```python
import pandas as pd
import seaborn as sns
```

```python
cars = pd.read_csv('data/Cars93.csv')
```

```python
cars = cars.loc[:,('Type','Price')]
```

```python
cars_type = cars.groupby('Type').count().reset_index()
```

```python
cars_type.columns = ('Type', 'Count')
```

```python
cars_type
```

```python
sns.barplot(data = cars_type, x='Type', y='Count')
```

#### (2) 상대도수분포표(relative frequency distribution table)

- 상대도수 : 구간별 도수를 전체 도수의 총합으로 나눈값

```python
cars.count()
```

```python
rfdt = (cars.groupby('Type').count()/cars['Type'].count()).reset_index()
rfdt #전체 합은 1
```
