### regplot

```python
automobile[['symboling','length']].head()
```

```python
from numpy import corrcoef
```

corrcoef : 상관관계

```python
corrcoef(automobile.symboling, automobile.length)
```

약한 음의 상관관계를 띠고 있다

```python
sb.pairplot(automobile, vars=['length','symboling'])
```

```python
sb.regplot(automobile.length,automobile.symboling)
```

```python
sb.regplot('length','symboling', data=automobile)
```

#### 회귀선 지우기

```python
sb.regplot('length','symboling',data=automobile, fit_reg= False)
```

#### Jitter (겹치는 점을 조금씩 엇갈리게 표현)

```python
sb.regplot('length','symboling',data=automobile, fit_reg=False, y_jitter=0.1)
```

```python
sb.regplot('length','symboling',data=automobile,y_jitter=0.1)
```

#### 신뢰구간 지우기

```python
sb.regplot('length','symboling',data=automobile,y_jitter=0.1,ci=None)
```

#### 색

```python
sb.regplot('length','symboling',data=automobile,y_jitter=0.1,ci=None,color='red')
```

#### 점 색

```python
sb.regplot('length','symboling',data=automobile,y_jitter=0.1,ci=None,scatter_kws={'color':'orange'})
```

#### 선 색

```python
sb.regplot('length','symboling',data=automobile,y_jitter=0.1,ci=None,line_kws={'color':'orange'})
```
